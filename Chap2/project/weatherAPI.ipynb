```python
import re, string, requests, os, json
from collections import namedtuple


class CityWeather(object):
    
    # 初始函数
    def __init__(self):
        self.weatherDict = {} # txt 转换成 dict
        self.ursCity = '' # 用户搜索的城市
        self.result = []  # 城市天气
        self.allResult = [] # 用户查询记录
        self.ursTxt = [] 
        self.theCity = False

        
     # 打印说明
    def printExplain(self):
         print("""* 这是一个天气查询系统，能够查询中国市地级的城市天气。
* 输入城市名称（支持中文、全拼），即可查询到您所需要的天气。
* 输入 h 或 help 得到帮助。\n
天气数据来源：心知天气 \n""")
        
    # 打印帮助
    def printHelp(self):
         print("""中国城市天气查询系统说明:
1. 输入城市名，返回该城市的天气数据；
2. 输入 h 或 help 指令，打印帮助文档；
3. 输入 history 指令，打印查询记录；
4. 输入 q 或 quit 指令，退出程序的交互。\n
        """)
       
        
    
    # 用户输入并写入记录
    def cityInput(self):
        while True:
            City = input("请输入您想查询的城市：")
            city = City.strip()
            count = self.str_count(city)
            if count.digit != 0:
                print("请输入中、英文字符！不要有数字！\n")
                continue
            elif count.space != 0:
                print("请去掉空格！\n")
                continue
            elif count.punc != 0:
                print("请去掉标点符号！\n")
                continue   
            else:
                self.usrCity = city
                return city
                      
     
    # 找出字符串中的中英文、空格、数字、标点符号，并统计个数
    def str_count(self, city):
        count_en = count_dg = count_sp = count_zh = count_pu = 0
        s_len = len(city)
        for c in city:
            if c in string.ascii_letters:
                count_en += 1
            elif c.isdigit():
                count_dg += 1
            elif c.isspace():
                count_sp += 1
            elif c.isalpha():
                count_zh += 1
            else:
                count_pu += 1
        total_chars = count_zh + count_en + count_sp + count_dg + count_pu
        if total_chars == s_len:
            return namedtuple('Count', ['total', 'zh', 'en', 'space', 'digit', 'punc'])(s_len, count_zh, count_en, count_sp, count_dg, count_pu)
        else:
            print('Something is wrong!')
            return None
        return None
    
    # 查询天气
    def weather(self, city):
        location = city
        personalKey = 'aujwvfmqnvpkrgrq'
        url = 'https://api.seniverse.com/v3/weather/now.json?key=' + personalKey  + '&location=' + location  + '&language=zh-Hans&unit=c'
        data = requests.get(url)
        import json
        json = json.loads(data.text)
        
        if 'status' in json and location != 'q' and location != 'quit' and location != 'h' and location != 'help' and location != 'history':
            print("您查询的城市不在查询库中！\n")
        elif location == 'q' or location == 'quit':
            json.get(location)
            print("退出查询系统")
        elif location == 'h' or location == 'help':
            json.get(location)
            print('')
        elif location == 'history':
            print('')
        else:    
            c = json["results"][0]["location"]["name"]
            w = json["results"][0]["now"]["text"]
            t = json["results"][0]["now"]["temperature"]
            updateTime = json["results"][0]["last_update"]
            
            collectCity = c 
            collect = c +', ' + w + ', ' + t + '℃' + ', ' + '更新时间：' + updateTime
         
            print('城市:', c)
            print('天气:', w)
            print('温度:', t)
            print('数据更新时间:', updateTime, '\n')
            
            self.result = collect
            self.allResult.append(collect)

    
        
    
    # 写入文件
    def writeFile(self, allResult):
        A = self.allResult
        S = set(A)
        L = list(S)
        File = open("record.txt", 'a')
        for i in S:
            File.write('%s %s' % (i, os.linesep))
        File.close() 
        rFile = open("/users/lixun/Python/Py101-004/Chap2/Project/record.txt", 'r')
        wFile = open("/users/lixun/Python/Py101-004/Chap2/Project/wRecord.txt", 'w')
        allLine = rFile.readlines()
        rFile.close()
        s = set()
        for k in allLine:
            s.add(k)
        for j in s:
            wFile.write('%s %s' % (j, os.linesep))
        wFile.close()
        wFile = open("/users/lixun/Python/Py101-004/Chap2/Project/wRecord.txt", 'r')
        R = wFile.read()
        self.ursTxt = R
        
    def printHistory(self, ursTxt):
        print('\n')
        print(self.ursTxt)
        

        
    # 运行
    def run(self):
        self.printExplain()
        while not self.theCity:
            try:
                city = self.cityInput()
                self.weather(city)
                if city == 'h' or city == 'help':
                    self.printHelp()
                if city == 'history':
                    print("\n您的查询记录为：")
                    self.writeFile(self.allResult)
                    self.printHistory(self.ursTxt)
                if city == 'q' or city == 'quit':
                    self.writeFile(self.allResult)
                    break
            except KeyboardInterrupt:
                print("Try again!")
       
              
# 主函数
def main():
    cityWeather = CityWeather()
    cityWeather.run()   
if __name__ == '__main__':
    main()
        
```
