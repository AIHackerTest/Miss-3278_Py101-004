#### 基本用法 requests.get()

```python
import requests

r = requests.get('https://api.github.com/events')

print(r.url)
```

```python
import requests

payload = {'key1': 'value1', 'key2': 'value2'}
r = requests.get('https://api.github.com/events',params=payload)

print(r.url)
```

```python
import requests

payload = {'key1': 'value1', 'key2': ['value2', 'value3']}
r  = requests.get('https://api.github.com/events',params=payload)

print(r.url)                 
```

#### requests.post()

```python
import requests, json

payload = {'key1': 'value1', 'key2': ['value2', 'value3']}
r = requests.post("http://httpbin.org/post", data = json.dumps(payload))

print(r.url)
```

```python
import requests

payload = {'key': 'value'}
r = requests.post("http://httpbin.org/post", json=payload)
```

```python
# coding=utf-8
import requests, json

location = input("> ")
personalKey = 'aujwvfmqnvpkrgrq'
url = 'https://api.seniverse.com/v3/weather/now.json?key=' + personalKey  + '&location=' + location  + '&language=zh-Hans&unit=c'
data = requests.get(url)
json = json.loads(data.text)



if 'status' in json and location != 'q' and location != 'quit' and location != 'h' and location != 'help':
    print("您查询的城市不在查询库中！")
elif location == 'q' or location == 'quit':
    json.get(location)
    print('quit')
elif location == 'h' or location == 'help':
    json.get(location)
    print('ok')
else:
    c = json["results"][0]["location"]["name"]
    w = json["results"][0]["now"]["text"]
    t = json["results"][0]["now"]["temperature"]
    updateTime = json["results"][0]["last_update"]

    collect = c +', ' + w + ', ' + t + '℃' + ', ' + '更新时间：' + updateTime
    
    print('城市:', c)
    print('天气:', w)
    print('温度:', t)
    print('数据更新时间:', updateTime)
```
