 做法：仅仅是看和理解，没有输入代码。ex43 主要讲了构建一个小程序的思路和做法（作者写了一个小游戏），这给我完成课程任务提供了参照做法。

  1. 写应用程序的思路：
  
    1）先想这个应用程序会用到的名词，罗列出来，然后做合并同类项或删除处理。这样方便构建类、函数或者变量的名称。
    
    2) 列出基本名词会使用的动词，罗列出来，就是程序要做的事儿，可以写成函数。

  2. 写应用程序的两条路径：
  
    1）自顶而下：先构建大框，然后添加细节。
    
    2）自下而上：先写某一块有灵感的或者熟悉的部分，然后构建程序架构。（我发现自己用的是这种方法，第一种则不会写了，不知道改写什么。）
