```python
import re, string, requests, os
from collections import namedtuple


class CityWeather(object):
    
    # 初始函数
    def __init__(self):
        self.weatherDict = {} # txt 转换成 dict
        self.ursCity = '' # 用户搜索的城市
        self.result = []  # 城市天气
        self.allResult = [] # 用户查询记录
        self.ursTxt = [] 
        self.theCity = False

        
     # 打印说明
    def printExplain(self):
         print("""* 这是一个天气查询系统，能够查询中国市地级的城市天气。
* 输入城市名称，即可查询到您所需要的天气。
* 输入 h 或 help 得到帮助\n""")
        
    # 打印帮助
    def printHelp(self):
         print("""中国城市天气查询系统说明:
1. 输入城市名，返回该城市的天气数据；
2. 输入 h 或 help 指令，打印帮助文档；
3. 输入 q 或 quit 指令，退出程序的交互；
4. 在退出程序之前，打印查询过的所有城市。\n
        """)
       
        
    # 生成字典
    def txt_Dict(self):
        weather = {}
        f = open('/Users/lixun/Python/Py101-004/Chap1/resource/weather_info.txt', 'r') 
        for line in f:
            v = line.strip().split(',')
            weather[v[0]] = v[1]
        self.weatherDict = weather
        
    
    # 用户输入并写入记录
    def cityInput(self):
        while True:
            City = input("请输入您想查询的城市：")
            city = City.strip()
            count = self.str_count(city)
            if (count.en != 0 or count.digit != 0) and City != 'q' and City != 'quit' and City != 'h' and City != 'help':
                print("请输入中文字符！不要有英文字符、数字！\n")
                continue
            elif count.space != 0:
                print("请去掉空格！\n")
                continue
            elif count.punc != 0:
                print("请去掉标点符号！\n")
                continue   
            else:
                self.usrCity = city
                return city
                      
     
    # 找出字符串中的中英文、空格、数字、标点符号，并统计个数
    def str_count(self, city):
        count_en = count_dg = count_sp = count_zh = count_pu = 0
        s_len = len(city)
        for c in city:
            if c in string.ascii_letters:
                count_en += 1
            elif c.isdigit():
                count_dg += 1
            elif c.isspace():
                count_sp += 1
            elif c.isalpha():
                count_zh += 1
            else:
                count_pu += 1
        total_chars = count_zh + count_en + count_sp + count_dg + count_pu
        if total_chars == s_len:
            return namedtuple('Count', ['total', 'zh', 'en', 'space', 'digit', 'punc'])(s_len, count_zh, count_en, count_sp, count_dg, count_pu)
        else:
            print('Something is wrong!')
            return None
        return None
    
    
    # 查询结果
    def showResult(self,city, weatherDict):
        if city in weatherDict:
            r = str(city)
            t = weatherDict[r]
            s = r + ' --> ' + t
            print(s)
            print('\n')
            self.result = t
            self.allResult.append(s)
        elif city == 'q' or city == 'quit':
            return
        elif city == 'h' or city == 'help':
            print('')
        else:
            print("您输入的城市不在查询库中。\n")
        
    
    # 写入文件
    def writeFile(self, allResult):
        A = self.allResult
        S = set(A)
        L = list(S)
        File = open("urs.txt", 'a')
        for i in S:
            File.write('%s %s' % (i, os.linesep))
        File.close() 
        rFile = open("/users/lixun/Python/Py101-004/Chap1/Project/urs.txt", 'r')
        wFile = open("/users/lixun/Python/Py101-004/Chap1/Project/wurs.txt", 'w')
        allLine = rFile.readlines()
        rFile.close()
        s = set()
        for k in allLine:
            s.add(k)
        for j in s:
            wFile.write('%s %s' % (j, os.linesep))
        wFile.close()
        wFile = open("/users/lixun/Python/Py101-004/Chap1/Project/wurs.txt", 'r')
        R = wFile.read()
        print('\n')
        print(R)
        self.ursTxt = R
        

        
    # 运行
    def run(self):
        self.txt_Dict()
        self.printExplain()
        while not self.theCity:
            city = self.cityInput()
            self.showResult(city, self.weatherDict)
            if city == 'h' or city == 'help':
                self.printHelp()
            if city == 'q' or city == 'quit':
                self.writeFile(self.allResult)
                break
       
              
# 主函数
def main():
    cityWeather = CityWeather()
    cityWeather.run()   
if __name__ == '__main__':
    main()
        
```
